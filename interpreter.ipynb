```python
import numpy as np
import pandas as pd
import h5py
import os
import csv
```

```python
os.chdir('TheBeatlesHDF5 2')
```

```python
albums = os.listdir()

for album in albums:
    if ((album[0] == '1') | (album[0] == '0')):
        continue
    else:
        albums.remove(album)
```

```python
md_df = pd.DataFrame()
album_df = pd.DataFrame()

names = []
i = 0

for album in albums:
    os.chdir(album)
    
    songs = os.listdir()
    num_tracks = len(songs)
    album = album[5:]
    album = album.replace('_', ' ').strip()
    
    if (album not in names):
        i = i + 1
        names.append(album)
    else:
        os.chdir("..")
        continue
        
    if (album[0] == '-'):
        album = album[2:]
    
    album_df = album_df.append([[i, album, "The Beatles", num_tracks]])
    
    for song in songs:
        songdir = os.getcwd() + '/' + song
        with h5py.File(songdir, 'r') as hdf:
            metadata = np.array(hdf.get('metadata').get('songs'))
            analysis = np.array(hdf.get('analysis').get('songs'))
            mb       = np.array(hdf.get('musicbrainz').get('artist_mbtags'))
            
            s_id  = metadata['song_id'][0]
            title = metadata['title'][0]
            duration = analysis['duration'][0]
            tempo = analysis['tempo'][0]
                        
            md_df = md_df.append([[s_id, title, duration, tempo, i]])
            
    os.chdir("..")
```

```python
md_df.columns    = ['s_id', 'title', 'duration', 'tempo', 'a_id']
album_df.columns = ['a_id', 'name', 'artist', 'tracks']
```

```python
with open('albums.csv', 'w', newline = '') as f:
    writer = csv.writer(f)
    
    writer.writerow(list(album_df.columns))
    
    for index, row in album_df.iterrows():
        writer.writerow((list(row)))
```

```python
with open('songs.csv', 'w', newline = '') as f:
    writer = csv.writer(f)
    
    writer.writerow(list(md_df.columns))
    
    for index, row in md_df.iterrows():
        title = str(row['title'])
        title = title[2:].strip()
        row.loc['title'] = title[:-1]
        
        sid = str(row['s_id'])
        sid = sid[2:].strip()
        row.loc['s_id'] = sid[:-1]
        
        writer.writerow((list(row)))
```
